```python
#Q.2 Given two strings word1 and word2, return the minimum number of operations required to convert word1to word2.
def distance(word1, word2):
    if not word1:
        return len(word2)
    if not word2:
        return len(word1)
    if word1[0] == word2[0]:
        return distance(word1[1:],word2[1:])
    else:
        return 1+min(
            distance(word1[1:],word2),#insert
            distance(word1,word2[1:]),#remove
            distance(word1[1:],word2[1:])#replace
        )
#DRIVER CODE 
r = "intention"
p = "execution"
distance(r,p) 
```

```python
#Q. 3 Print the max value of the array [ 13, 1, -3, 22, 5]
def find_max(array,n):
    # FIRST WE TAKE BASE CASE CONDITIONE
    if(n==1):
        return array[0]
    else:
        #now we doing recursion over here 
        return max(array[n-1],find_max(array,n-1))
    
array = [ 13, 1, -3, 22, 5]
n = len(array)
find_max(array,n)    
#time complexity O(n)
#space complexity O(n)
#T(n) = T(n-1)+c
#T(n) = T(n-2)+2c
#T(n) = T(n-3)+3c
#T(n) = T(n-k)+kc
#(n-k)= 1 base case conditione
#T(n) = T(n-n+1)+n.c
#T(n) = 1+n.c
```

```python
#Q.4 Find the sum of the values of the array [92, 23, 15, -20, 10]
def sum_of_array(array,n):
    if array == []:
        return "I AM EXCPECTING SOME DATA"
    if n<=0:
        return 0
    else:
        return sum_of_array(array,n-1)+array[n-1]
    
#driver code
array = [92, 23, 15, -20, 10]
n = len(array)
sum_of_array(array,n)
#time complexity O(n)
#space complexity O(n)
#T(n) = T(n-1)+c
#T(n) = T(n-2)+2c
#T(n) = T(n-3)+3c
#T(n) = T(n-k)+kc
#(n-k)= 0 base case conditione
#T(n) = T(n-n)+n.c
#T(n) = 0+n.c
```

```python
#Q.5 Given a number n. Print if it is an armstrong number or not.An armstrong number is a number if the sum 
#of every digit in that number raised to the power of total digits in that number is equal to the number
def digit(n,i):
    if i == 1:
        dig = n % 10
    else:
        a = n // 10**(i-1)
        dig = a % 10
    return(dig)
def is_armstorng(n,i,Y):
    Z = len(str(n))
    while i<=len(str(n)):
        x = digit(n,i)
        N = x**Z
        Y = Y + N
        return is_armstorng(n,i+1,Y)
    if Y== n:
        print("YES!")
    else:
        print("NO")
print("INPUT 'i== 1','Y==0'AND ENTER N == 'YOUR DIGIT'")        
```
